```python
para finalizar haremos un testeo de hipotesis. realizamos una tabla pivot para juntar nuestras dos variables para hacer el testeo.
sera que habra una relacion entre la especialidad con repecto de la edad de los pacientes.
```

```python
cross = pd.crosstab(new['Specialty_Name'], new['Age_Profile'])

cross
```

para saber la respuesta de nuestra hipotesis utilizaremos el chi cuadrado para determinar dicha correlacion.
nuestra hipotesis es que H1 = no hay una diferencia de la edad del pacinete con respecto de la especialidad que visita. 

```python
from scipy.stats import chi2_contingency


estadistico, p_valor, ex, ddof = chi2_contingency(cross)

estadistico, p_valor
```

En este caso, con un valor p de 0.0, significa que la probabilidad de observar resultados tan extremos (o más extremos)
bajo la suposición de que no hay relación entre las variables (H0 es verdadera)es extremadamente baja, prácticamente nula.

```python
from scipy.stats.contingency import association

association(cross, method='cramer')
```

El valor de Cramer's V sugiere una asociación moderada entre las variables categóricas en tu análisis.
Esta asociación no es ni muy débil ni muy fuerte, sino que se encuentra en un nivel moderado.
Indica que hay una relación entre las variables, pero esta relación no es extremadamente fuerte ni determinante.

```python
age = pd.crosstab(con['Time_Bands'], con['Age_Profile'])

age
```

la siguiente hipotesis tiene que ver entre el tiempo de espera con respecto al rango de edad especifico.
H0 = no hay una diferencia
H1 = si se encuentra una diferencia

```python
estadistico, p_valor, ex, ddof = chi2_contingency(age)

estadistico, p_valor
```

indica que la probabilidad de obtener una correlación tan fuerte entre las variables bajo la hipótesis nula es esencialmente nula. Por lo tanto, se rechaza la hipótesis nula.

```python
association(age, method='cramer')
```

esto es lo que nos dice es testeo por cramer Este valor de 0.0264 indica una asociación muy débil entre las variables categóricas en tu análisis.
