Luego de subir todas las tablas procedemos a concatenarlas para obtener una tabla con todos los datos intengrados en una sola.

```python
concatenated = pd.concat([op_2018, op_2019, op_2020, op_2021])

concatenated
```

```python
Procedemos a la limpieza de datos, en este proceso se visealiza los datos, se aseguran que todos esten ordenados y limpios.
Además se eliminan valores nula, se borran columnas repetidas.
```

```python
concatenated.Time_Bands.unique()
```

```python
concatenated.Age_Profile.unique()
```

```python
value_to_replace = ' 0-15'
new_value = '0-15'
concatenated['Age_Profile'] = concatenated['Age_Profile'].replace(value_to_replace, new_value)
```

```python
concatenated.Age_Profile.unique()
```

```python
concatenated.isna().sum()
```

leugo de concatenar miramos si queda algun nulo por quitar de la nueva tabla.

```python
new = concatenated.dropna()

new
```

```python
change_value = lambda x: '18+ Months' if x == '18 Months +' else x
```

```python
new['Time_Bands'] = new['Time_Bands'].apply(change_value)

new['Time_Bands'].unique()
```

```python
procedemos a un testeo de hipotesis para demoistrar si se encunetra alguna correlación entre las variables.
```

```python
cross = pd.crosstab(new['Age_Profile'], new['Time_Bands'])

cross
```

```python
H0 = no ahy ninguna realcion entre si
H1 = si se encuentra una relación
```

```python
from scipy.stats import chi2_contingency


estadistico, p_valor, ex, ddof = chi2_contingency(cross)

estadistico, p_valor
```

```python
este valor indica que la probabilidad de observar estos resultados bajo la hipótesis nula es prácticamente nula.
Por lo tanto, se concluye que hay una diferencia significativa entre los grupos comparados.
```

```python
from scipy.stats.contingency import association

association(cross, method='cramer')
```

Dado que el valor de Cramer's V es relativamente bajo, indica que la asociación entre las variables categóricas es débil y no muy significativa. No proporciona una evidencia sólida de una diferencia significativa entre los grupos comparados.

```python
special = pd.crosstab(new['Age_Profile'], new['Speciality'])

special
```

```python
ahora comparemos el grupo de edad por especialidad.
```

```python
from scipy.stats import chi2_contingency


estadistico, p_valor, ex, ddof = chi2_contingency(special)

estadistico, p_valor
```

El valor p es la probabilidad de observar los resultados obtenidos, o resultados más extremos, si la hipótesis nula fuera verdadera.
Un valor p de 0.0 indica que la probabilidad de observar estos resultados bajo la hipótesis nula es esencialmente nula.

```python
from scipy.stats.contingency import association

association(special, method='cramer')
```

Este valor de 0.2727 indica una asociación moderada entre las variables categóricas en tu análisis.
A diferencia de los valores anteriores, este valor es más sustancial y sugiere una asociación más fuerte entre las variables.
